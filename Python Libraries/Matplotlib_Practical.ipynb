# Legend, Grid

```python
Day = [1,2,3,4,5,6]
min_T = [10,11,19,13,10,18]
max_T = [20,23,29,26,20,28]
avg_T = [30,31,39,33,30,40]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(Day, min_T, label="min")
plt.plot(Day, max_T, label="max")
plt.plot(Day, avg_T, label="avg")

plt.legend(loc="best", shadow="True", fontsize="small")

plt.grid()
```

# Barchart :-

```python
import numpy as np
```

```python
company = ["Google", "Facebook", "Instragram", "Wipro"]
revenue = [90,136,89,30]
profit = [40,5,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel("Revenue")
plt.title("US Tech Stocks")

plt.bar(xpos-0.2, revenue, label="Revenue", width=0.4)
plt.bar(xpos+0.2, profit, label="Profit", width=0.4)
plt.legend(loc="best",shadow="True")
```
